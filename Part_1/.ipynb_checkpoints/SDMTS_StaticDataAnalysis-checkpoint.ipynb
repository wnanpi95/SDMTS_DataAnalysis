```r
library(ggplot2)
library(dplyr)
library(ggmap)
library(maps)
library(mapdata)
library(lubridate)
options(jupyter.plot_mimetypes = 'image/png')
```

```r
path <- "google_transit"
extension <- ".txt"
files <- c("agency", "calendar_dates", "calendar", "fare_attributes", "fare_rules",
           "feed_info", "frequencies", "places", "routes", "shapes", "stop_times", "stops",
           "transfers", "trips")

GTFSlist <- list()

for (i in 1:length(files)) {
  df <- read.csv(file=file.path(path, paste(files[i], extension, sep="")), header=T)
  GTFSlist[[i]] <- df
}
```
