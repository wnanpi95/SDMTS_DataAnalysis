```r
library(ggplot2)
library(dplyr)
library(ggmap)
library(maps)
library(mapdata)
library(lubridate)
options(jupyter.plot_mimetypes = 'image/png')
```

```r
real_time_df <- read.csv("./real_time_data_parsed.csv", header=T)
head(real_time_df)
```

A look at the positions of all the buses at this time slice

```r
size_bbox <- 0.2
bbox <- make_bbox(lon = real_time_df$longitude, lat = real_time_df$latitude, f = size_bbox)
```

```r
sq_map <- get_map(location = bbox, maptype = "roadmap", source = "google")
ggmap(sq_map) + geom_point(data = real_time_df,
                           mapping = aes(x = longitude, y = latitude), 
                           color = "red", alpha = 1.0, size = 0.3)
```
